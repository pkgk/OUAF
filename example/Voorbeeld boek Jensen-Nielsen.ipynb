#### Arjen algemeeen
Om te beginnen met de terminologie: vroeger werden Bayesiaanse netwerken ook wel ‘Bayesian belief networks’ genoemd. Het idee daarbij is dat een agent zijn geloof (“belief”) in allerlei variabelen vastlegt door middel van een B(B)N. Je hebt daarbij de a priori beliefs, bijv. dat P(W = wet) = 0.3, maar nadat je een forecast van rainy hebt geobserveerd, dan veranderen de beliefs van de agent.

#### Arjen: er staat dus in de eerste rij van tabel 4.1:
- Bel(W = wet) = P(W = wet | F = rainy) = P(F=rainy | W=wet) P(W=wet) / P(F=rainy) = (0.6 * 0.3) / (0.6 * 0.3 + 0.1 * 0.7) = 0.72


- Bel(R = melb_wins) = 
- P(R = melb_wins | F = rainy) = 
- P(R = melb_wins | W=wet) P(W = wet | F = rainy) + P(R = melb_wins | W=dry) P(W = dry | F = rainy) = 
- 0.6 * 0.72 + 0.25 * (1-0.72) = 0.502

#### Arjen: mbt het besluit
De information links geven welke informatie bekend is bij de agent voordat hij besluit welke beslissing hij neemt. Dat is anders dan Figure 4.3 waarbij de beslissing niet van een andere variabele afhangt. De beslissing “Accept bet” is dus een functie van Forecast in Figure 4.4. Dat zie in die Decision Table bij Figure 4.4 waarbij bij verschillende F een verschillende beslissing wordt genomen (dit heet ook wel de ‘policy’).

Nu terug naar Tabel 4.1: hier wordt bepaald wat de optimale policy is. Er wordt in de eerste rij berekend wat de expected utility is voor bijv. AB=yes in de situatie waarbij F = rainy:

- EU(AB=yes | F=rainy) = 
- P(R = melb_wins | F = rainy) * U(R = melb_wins | AB = yes) + P(R = melb_loses | F = rainy) * U(R = melb_loses | AB = yes) = 
- 0.502 * 40 + (1-0.502) * -20 = 10.12.

Dit is blijkbaar groter dan EU(AB=no | F=rainy), dus als de agent de forecast rainy heeft gezien dan zou hij de bet moeten accepteren. In de decision table van Figure 4.4 staat de optimale policy die in Tabel 4.1 wordt berekend (dat is dus niet ‘gegeven’, dat is denk ik de verwarring).
