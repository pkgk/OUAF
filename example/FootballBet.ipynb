### creeer influence diagram:

```python
# maak diagram instantie
diag = gum.InfluenceDiagram()

# voeg kans nodes toe en arc ertussen
W=diag.addChanceNode(gum.LabelizedVariable("W","W",2))
R=diag.addChanceNode(gum.LabelizedVariable("R","R",2))
diag.addArc(diag.idFromName("W"),R)

# voeg decision en utility nodes to en links
D=diag.addDecisionNode(gum.LabelizedVariable("D","D",1))
U=diag.addUtilityNode(gum.LabelizedVariable("U","U",1))
diag.addArc(diag.idFromName("R"),U)
diag.addArc(diag.idFromName("D"),U)

# laat diagram zien
gnb.showInfluenceDiagram(diag)
```

```python
diag.cpt(W).fillWith([0.3,0.7])
diag.cpt(R)[{'R':0}] =[0.3,0.7]
diag.cpt(R)[{'R':1}] =[0.6,0.25]



#diag.cpt(W)[{'W':0}]=[3]

#diag.utility(U)[{'F':0,'decisionVar3':0}]=2
#diag.utility(U)[{'F':0,'decisionVar3':1}]=4
#diag.utility(U)[{'F':1}]=[[0],[5]]
```
